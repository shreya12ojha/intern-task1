```python
#Create 2 simple plots: 1 bar chart & 1 histogram or line plot
plt.figure(figsize=(8,4))
df["category"].value_counts().plot(kind="bar")
plt.title("Ticket count by category")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
df["created_date"] = pd.to_datetime(df["created_date"])
daily_counts = df.groupby("created_date")["ticket_id"].count()
plt.figure(figsize=(8,4))
daily_counts.plot()
plt.title("Tickets over time")
plt.xlabel("Date")
plt.ylabel("Number of tickets")
plt.show()
```

SUMMARY

This dataset contains a few hundred customer support tickets from multiple regions, channels, and priorities, with timestamps in early 2025.

Each row represents a single ticket with its resolution time, agent’s experience, category, and customer satisfaction flag.

Bug Report, Login/Access, Billing, and Data Issue appear as the main categories, with Bug Report typically being the most frequent.

Resolution times are quite spread out, with many tickets resolved relatively quickly but a noticeable tail of tickets that take much longer than the median.

Average resolution time varies by priority and category; some high or urgent tickets take substantially longer, suggesting higher complexity.

Satisfaction tends to be higher for tickets with shorter resolution times and for certain categories or channels.

There are missing values in fields like agent_experience_years, region, and channel, which may affect some analyses.

A few tickets have missing resolution times, so these should be handled carefully for any modeling or analysis.
